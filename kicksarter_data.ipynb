```python
# We would consider someone as experiencing the most severe form at that time
def clean_homelessness_form(val):
    if isinstance(val, str):
        v = val.lower()
        if "primary and secondary" in v:
            return "Primary"
        elif "secondary and tertiary" in v:
            return "Secondary"
        else:
            # Remove any accidental whitespace/casing
            return val.strip().title()
    else:
        return "No"

# df_zapier['RF - Form of homelessness (Resolved)'] = df_zapier['RF - Form of homelessness Experienced (Primary, Secondary and Teritary)'].apply(clean_homelessness_form)
df_zapier['RF - Currently experiencing (Resolved)'] = df_zapier['RF - Currently experiencing'].apply(clean_homelessness_form)
print(df_zapier['RF - Form of homelessness Experienced (Primary, Secondary and Teritary)'].value_counts())
df_zapier['RF - Currently experiencing (Resolved)'].value_counts()
```

```python
df_zapier['RF - Form of homelessness Experienced (Primary, Secondary and Teritary)'] = (
    df_zapier['RF - Form of homelessness Experienced (Primary, Secondary and Teritary)']
    .str.strip()
    .replace({'Seconday': 'Secondary'})
)
df_zapier['RF - Form of homelessness Experienced (Primary, Secondary and Teritary)'].value_counts()
```

```python
def print_value_counts(df, name):
    print(f"\n===== Value Counts for {name} =====\n")
    for col in df.columns:
        print(f"--- {col} ---")
        print(df[col].value_counts(dropna=False))
        print("\n")

# For each dataframe:
print_value_counts(df_zapier, "Zapier Data")
print_value_counts(df_sm3, "SurveyMonkey 3 Months Data")
print_value_counts(df_sm6, "SurveyMonkey 6 Months Data")
print_value_counts(df_id, "ID Data")
```

```python
# define age bins
age_bins = [0, 18, 25, 35, 50, 120]
age_labels = ['<18', '18-24', '25-34', '35-49', '50+']
df_zapier['Age Group'] = pd.cut(df_zapier['Age'], bins=age_bins, labels=age_labels, right=False)
```

## Merge and Export

```python
# add prefix to columns
def prefix_columns(df, prefix):
    df = df.copy()
    df.columns = [prefix + col if col != 'IDENTIFIER' else col for col in df.columns]
    return df

df_zapier_prefixed = prefix_columns(df_zapier, 'ZAPIER_')
df_sm3_prefixed    = prefix_columns(df_sm3, 'SM3_')
df_sm6_prefixed    = prefix_columns(df_sm6, 'SM6_')
```

```python
# Merge
df_merged = df_id.copy()
df_merged = df_merged.merge(df_zapier_prefixed, on='IDENTIFIER', how='outer')
df_merged = df_merged.merge(df_sm3_prefixed, on='IDENTIFIER', how='outer')
df_merged = df_merged.merge(df_sm6_prefixed, on='IDENTIFIER', how='outer')
```

```python
# Export individual cleaned data
df_zapier.to_csv('data/clean_Internal_Zapier_Data.csv', index=False)
df_sm3.to_csv('data/clean_Internal_Surveymonkey_3_Months_Data.csv', index=False)
df_sm6.to_csv('data/clean_Internal_Surveymonkey_6_Months_Data.csv', index=False)
df_id.to_csv('data/clean_Internal_ID_Data.csv', index=False)

# Export merged data with prefixes for dashboard use
df_merged.to_csv('data/clean_MERGED_All_Data.csv', index=False)
```
